```python
import pandas as pd
```

```python
df = pd.read_csv("data/weather_stations.csv", header=None, skiprows=2, delimiter=';')
```

```python
df.columns = ['station', 'measure']
```

```python
df = df.groupby('station')['measure'].agg(['min', 'max', 'mean']).reset_index().sort_values('station')
```

```python
df
```

```python
for i in range(len(df)):
    print(f"{df.loc[i].station}={round(df.loc[i]['min'],1)}/{round(df.loc[i]['mean'],1)}/{round(df.loc[i]['max'],1)}", end=', ')
```
